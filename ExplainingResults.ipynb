```python
import numpy as np
import irreducible_library as irr
import os, sys
import matplotlib.pyplot as plt
import scipy
%matplotlib inline
%load_ext autoreload
plt.style.use('default')
from matplotlib import rc, rcParams
from irreducible_library import relic, relic_ann, relic_plas
# Make use of TeX\ufeff
#rc('text',usetex=True)
# Change all fonts to ‘Computer Modern’
rc('font',**{'size':14, 'family':'serif','serif':['Times New Roman']})
```

```python
f, axs=plt.subplots(1,2)
f.set_figwidth(8)
Tlist = np.logspace(2.5,-2.5, num=200)
ax = axs[0]
ax.loglog(Tlist, [irr.gamma_ann(10e-12,0.04,i) for i in Tlist],label='$Q=10e-12$')
ax.loglog(Tlist, [irr.gamma_ann(10e-13,0.04,i) for i in Tlist],label='$Q=10e-13$')
ax.loglog(Tlist, [irr.gamma_ann(10e-14,0.04,i) for i in Tlist],label='$Q=10e-14$')
ax.set_xlabel('$T$ [MeV]',fontsize=20)
ax.set_ylabel('$\Gamma_{ann}$',fontsize=20)
ax.set_title("$m_\chi=40$keV")
ax.set_xlim([10,0.005])
ax.legend()

ax = axs[1]
ax.loglog(Tlist, [irr.gamma_ann(10e-12,0.04,i) for i in Tlist],label='$m_\chi=40$keV')
ax.loglog(Tlist, [irr.gamma_ann(10e-12,0.004,i) for i in Tlist],label='$m_\chi=4$keV')
ax.loglog(Tlist, [irr.gamma_ann(10e-12,0.00004,i) for i in Tlist],label='$m_\chi=40$eV')
ax.set_xlabel('$T$ [MeV]',fontsize=20)
ax.set_ylabel('$\Gamma_{ann}$',fontsize=20)
ax.set_title("$Q=10e-12$keV")
ax.set_xlim([10,0.005])
ax.set_ylim([10e-100,10e-23])
ax.legend()
f.tight_layout()
plt.show()
```

```python
f, axs=plt.subplots(1,2)
f.set_figwidth(8)
Tlist = np.logspace(2.5,-2.5, num=200)
ax = axs[0]
ax.loglog(Tlist, [irr.gamma_long(10e-12,0.04,i) for i in Tlist],label='$Q=10e-12$')
ax.loglog(Tlist, [irr.gamma_long(10e-13,0.04,i) for i in Tlist],label='$Q=10e-13$')
ax.loglog(Tlist, [irr.gamma_long(10e-14,0.04,i) for i in Tlist],label='$Q=10e-14$')
ax.set_xlabel('$T$ [MeV]',fontsize=20)
ax.set_ylabel('$\Gamma_{long}$',fontsize=20)
ax.set_title("$m_\chi=40$keV")
ax.set_xlim([100,0.5])
ax.legend()

ax = axs[1]
ax.loglog(Tlist, [irr.gamma_long(10e-12,0.04,i) for i in Tlist],label='$m_\chi=40$keV')
ax.loglog(Tlist, [irr.gamma_long(10e-12,0.004,i) for i in Tlist],label='$m_\chi=4$keV')
ax.loglog(Tlist, [irr.gamma_long(10e-12,0.00004,i) for i in Tlist],label='$m_\chi=40$eV')
ax.set_xlabel('$T$ [MeV]',fontsize=20)
ax.set_ylabel('$\Gamma_{long}$',fontsize=20)
ax.set_title("$Q=10e-12$keV")
ax.set_xlim([100,0.02])
ax.set_ylim([10e-50,10e-20])
ax.legend()
f.tight_layout()
plt.show()
```
