```python
import numpy as np
import irreducible_library as irr
import os, sys
import matplotlib.pyplot as plt
import scipy
%matplotlib inline
%load_ext autoreload
plt.style.use('default')
from matplotlib import rc, rcParams
from irreducible_library import relic, relic_ann, relic_plas
# Make use of TeX\ufeff
#rc('text',usetex=True)
# Change all fonts to ‘Computer Modern’
rc('font',**{'size':14, 'family':'serif','serif':['Times New Roman']})
```

```python
#mlist = np.logspace(0.001,1,num=200) # keV - MeV
mlist = np.logspace(-3,0,num=10)
Q = [10e-11,10e-12,10e-13,10e-14,10e-15]
a = irr.a_full(100) # a at T = 100 MeV
```

## Relic Abundance from Plasmon Decays

```python
Qm11 = np.array([relic_plas(Q[0],i)/irr.nxrelic(i, irr.T(100))*a**3/100**3 for i in mlist])
#4.41199726e-09
```

```python
Qm12 = np.array([relic_plas(Q[1],i)/irr.nxrelic(i, irr.T(100))*a**3/100**3 for i in mlist])
```

```python
Qm13 = np.array([relic_plas(Q[2],i)/irr.nxrelic(i, irr.T(100))*a**3/100**3 for i in mlist])
```

```python
Qm14 = np.array([relic_plas(Q[3],i)/irr.nxrelic(i, irr.T(100))*a**3/100**3 for i in mlist])
```

```python
Qm15 = np.array([relic_plas(Q[4],i)/irr.nxrelic(i, irr.T(100))*a**3/100**3 for i in mlist])
```

```python
f, ax = plt.subplots()

ax.loglog(mlist, [1 for i in mlist],lw=3,c='grey',alpha = 0.4,label = '$\Omega_\mathrm{DM}/\Omega_\mathrm{relic}$=1')
ax.loglog(mlist, Qm11,lw=2,c='SlateBlue',label = '$Q=10^{-11}$')
ax.loglog(mlist, Qm12,lw=2,c='lightcoral',label = '$Q=10^{-12}$')
ax.loglog(mlist, Qm13,lw=2,c='lightseagreen',label = '$Q=10^{-13}$')
ax.loglog(mlist, Qm14,lw=2,c='mediumorchid',label = '$Q=10^{-14}$')
ax.loglog(mlist, Qm15,lw=2,c='yellowgreen',label = '$Q=10^{-15}$')

ax.set_ylabel(r'$\Omega_\mathrm{DM}/\Omega_\mathrm{relic}$', fontsize=20)
ax.set_xlabel(r'$m_\chi$ [MeV]', fontsize=20)
ax.set_title('Relic Abundance from Plasmon Decays', fontsize=18)
#ax.set_xlim([20,0.1])
#ax.set_ylim([0.0006, 1.5])
#ax.set_xticklabels(xticklabels,fontsize=16)
ax.xaxis.set_tick_params(labelsize=16)
ax.yaxis.set_tick_params(labelsize=16)


ax.legend(loc='lower right',fontsize=10)
```

## Relic Abundance From Annihilations

```python
Qm11 = np.array([relic_ann(Q[0],i)/irr.nxrelic(i, irr.T(100))*a**3/100**3 for i in mlist])
Qm12 = np.array([relic_ann(Q[1],i)/irr.nxrelic(i, irr.T(100))*a**3/100**3 for i in mlist])
Qm13 = np.array([relic_ann(Q[2],i)/irr.nxrelic(i, irr.T(100))*a**3/100**3 for i in mlist])
Qm14 = np.array([relic_ann(Q[3],i)/irr.nxrelic(i, irr.T(100))*a**3/100**3 for i in mlist])
Qm15 = np.array([relic_ann(Q[4],i)/irr.nxrelic(i, irr.T(100))*a**3/100**3 for i in mlist])
```
